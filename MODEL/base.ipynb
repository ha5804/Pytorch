# Create SimpleLSTM

- SimpleLSTM을 생성하여 위의 데이터를 사용해보자.

**input_size는 feature 갯수이다.**

**hidden_size는 모델 표현력과 계산량을 고려하여 선택하는 하이퍼 파라미터이다.**

위 SimpleLSTM의 모델 크기(파라미터 수)는
`4 * (input_size + hidden_size) * hidden_size`
(4개의 gate로 연산량 4배)


````python
import torch
import torch.nn as nn


class SimpleLSTM(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers):
        super().__init__()

        self.lstm = nn.LSTM(
            input_size = 3,
            hidden_size = 32,
            num_layers = num_layers,
            batch_first = True
        )
    
    def forward(self, x):
        out, (h, c) = self.lstm(x)

        return out
````

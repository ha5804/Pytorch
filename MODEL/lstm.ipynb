### (2) data tensor 변환

- torch를 이용하여, lstm 모델에 사용할 데이터 tensor변환.
- torch.tensor(data[features_list])로 변환 가능.
- data shape 확인 중요 -> dtype 설정 필요.

```python
import torch

# x = torch.tensor(data[['Open', 'High', 'Low', 'Close', 'Volume']])
#error, dtype 오류

x = torch.tensor(data[['Open','High','Low','Close','Volume']].values,
                 dtype = torch.float32)

print(x.shape)
```

### (3)sequence data

- lstm은 시계열 데이터에 유리한 모델이다.
- sequence 처리는 필수는 아니지만, 거의 사용한다.

`window sliding`

```python
seq_length = 5

x_seq = []
y_seq = []

# x의 길이가 10라고 가정.
# 10 - 5 = 5
# i는 0 부터 5까지 반복.
for i in range(len(x) - seq_length):
    #i가 0이라면, 0부터 4까지 x에 추가
    #y에는 0~4일 데이터를 통한 5일 값 예측이므로,
    #정답값인 close(종가)이용.
    x_seq.append(x[i:i+seq_length])
    y_seq.append(x[i+seq_length][3])

x_seq = torch.stack(x_seq)
y_seq = torch.tensor(y_seq)
```
